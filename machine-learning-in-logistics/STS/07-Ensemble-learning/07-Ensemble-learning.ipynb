## 1 Ensemble Methods <a class="anchor" id="t1"></a>

First we build a normal decision tree using the code from exercise 3.

```python
# Apply the DecisionTreeClassifier algorithm
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
df = pd.read_csv('02-linear-regression_2.csv')

x = ['shiptype', 'Length', 'Breadth', 'SOG', 'TotalDistance', 'Duration']
y = ['Direct']

X_train, X_test, y_train, y_test = train_test_split(df[x], df[y], test_size = 0.2, random_state = 1)

dt = DecisionTreeClassifier(random_state = 1)
dt.fit(X_train, y_train)
predictions = dt.predict(X_test)
print(accuracy_score(predictions, y_test))
print(dt.get_depth())
print(dt.get_n_leaves())
```

You can later compare the result with the result of Bagging, Random forests and AdaBoost
